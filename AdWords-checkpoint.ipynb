```r
global_temp <- read.csv("C:/Users/Chris Iyer/Dropbox/Vendor Data - Import and Archive/Adwords/Adwords_2017-12-26.csv")
```

```r
library(dplyr)
library(lubridate)
library(lucr)
camp <- c("FY2018_Gmail","FY2018_YouTube","FY2018_Online","FY2018_Transfer" ,"FY2018_Undergrad" ,"FY2018_Courses" ,"FY2018_Graduate" , "FY2018_Doctoral","Gmail Test")
ad <- global_temp %>% filter(Campaign %in% camp)

ad$Day <- as.Date(ad$Day, format = "%d-%b-%y")
ad$Impressions <- as.numeric(ad$Impressions)

ad$Cost  <- from_currency(ad$Cost, decimal_delim = ".")
ad$Bounce.rate <- as.numeric(sub("%", "", ad$Bounce.rate))
ad <- ad %>% select(Date = Day, Campaign, Campaign.type, Ad.group, Impressions, Clicks, Cost, Bounce.rate) 

ad <- ad %>% mutate(CTR = round(Clicks/Impressions, digits = 3), CPC = round((Cost/Clicks), 2), CPM = round((Cost/Impressions)* 1000, digits = 2), Bounce.rate = round(Bounce.rate, 2))
head(ad, 20)
```
